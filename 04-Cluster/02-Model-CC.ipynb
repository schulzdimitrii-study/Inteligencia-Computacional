## ELBOW

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
# k is range of number of clusters.
visualizer = KElbowVisualizer(model, k=(2,15), timings= True)
visualizer.fit(X)        # Fit data to visualizer
visualizer.show()        # Finalize and render figure
```

### Escolhido 6 Clusters

```python
kmean= KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
kmean.fit(X)
labels=kmean.labels_
```

```python
clusters_labels = pd.DataFrame({'cluster':labels})
clusters_labels.head(3)
```

```python
clusters=pd.concat([df,clusters_labels], axis=1)
clusters.head()
```

## Interpretao dos Clusters

```python
import warnings
warnings.filterwarnings(action="ignore")
```

```python
for c in clusters:
    grid= sns.FacetGrid(clusters, col='cluster')
    grid.map(plt.hist, c)
```

# Visualizao dos Clusters, usando PCA

Usando PCA para transformar dados em 2 dimenses para visualizao

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
```

```python
dist = 1 - cosine_similarity(X)

pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)
X_PCA.shape
```

```python
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'blue',
          2: 'green', 
          3: 'yellow', 
          4: 'orange',  
          5:'purple'}

names = {   0:    'quem faz todos os tipos de compras',
            1:    'mais pessoas com pagamentos devidos',
            2:    'quem compra majoritariamente parcelado',
            3:    'quem recebe mais dinheiro adiantado',
            4:    'quem faz compras caras',
            5:    'quem no gasta muito dinheiro'
            }
  
df = pd.DataFrame({'x': x, 'y':y, 'label':labels}) 
groups = df.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend()
ax.set_title("Customers Segmentation based on their Credit Card usage bhaviour.")
plt.show()
```
