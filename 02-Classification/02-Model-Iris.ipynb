## Load Dataset

```python
df = pd.read_pickle(f"{ROOT}/data/iris.pickle")
df.head()
```

```python
X = df.drop('Target', axis=True)
X.head()
```

```python
y = df.Target
y.head()
```

```python
y.value_counts()
```

### Split

```python
from sklearn.model_selection import train_test_split   
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = SEED)   # Divide o dataset em treino e teste, mantendo a proporção das classes.
```

```python
y_train.value_counts()   # Verifica a distribuição das classes no conjunto de treino.
```

### Model Selection

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifiers = {
    "KNN" : KNeighborsClassifier(), # KNN com k=5 por padrão
    "KNN(3)" : KNeighborsClassifier(3), # KNN com k=3
    "DT" : DecisionTreeClassifier(), # Decision Tree sem limite de profundidade
    "DT(max_depth=5)" : DecisionTreeClassifier(max_depth=5), # Decision Tree com profundidade máxima de 5
    "RF" : RandomForestClassifier() # Random Forest (uma coleção de Decision Trees) sem limite de profundidade.
}
```

```python
from sklearn.metrics import accuracy_score # accuracy_score é uma métrica de avaliação de modelos de classificação.
from sklearn.preprocessing import StandardScaler # Padroniza os dados para que tenham média 0 e desvio padrão 1.

for name, model in classifiers.items():
    
    model.fit(X_train, y_train)
    
    # Scoring on SEEN data - effectively "useless" 
    y_pred = model.predict(X_train)       # predições (são as classes previstas pelo modelo) no conjunto de treino. 
    train_accuracy = accuracy_score(y_train, y_pred) # calcula a acurácia (proporção de acertos) no conjunto de treino.
    
    # Scoring on UNSEEN data - important
    y_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    print(f"{name:20s} accuracy\ttrain = {train_accuracy:.2%} \ttest = {test_accuracy:.2%}") # Exibe o nome do modelo, a acurácia no conjunto de treino e no conjunto de teste.
```

```python
scaler = StandardScaler() # Padroniza os dados para que tenham média 0 e desvio padrão 1.

X_train = scaler.fit_transform(X_train) # Aplica a padronização no conjunto de treino.
X_test = scaler.transform(X_test) # Aplica a padronização no conjunto de teste, usando os parâmetros do conjunto de treino.
```

```python
for name, model in classifiers.items():
    
    model.fit(X_train, y_train)
    
    # Scoring on SEEN data - effectively "useless"
    y_pred = model.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_pred)
    
    # Scoring on UNSEEN data - important
    y_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    print(f"{name:20s} accuracy\ttrain = {train_accuracy:.2%} \ttest = {test_accuracy:.2%}")
```
